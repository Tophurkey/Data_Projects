```python
import requests
import pandas as pd
from sqlalchemy import create_engine

#List of all six Current sources from https://data.cms.gov/medicare-shared-savings-program looking at "All Datasets"
cms_sources = ['Participants', 'SNF', 'ACO', 'Beneficiary', 'Performance', 'Risk']

#Create a new dict for storing all of our info and variables into keys for future use.
dict_sources = {}

dict_sources['Participants'] = {'url': "https://data.cms.gov/data-api/v1/dataset/5b678653-aa36-455b-9144-1d073ef7991b/data"}
dict_sources['SNF'] = {'url': "https://data.cms.gov/data-api/v1/dataset/6e053921-7f8d-44df-aa61-2654e0f35ef3/data"}
dict_sources['ACO'] = {'url': "https://data.cms.gov/data-api/v1/dataset/b9d03601-1187-4c25-aca6-5754e88ba0cf/data"}
dict_sources['Beneficiary'] = {'url': "https://data.cms.gov/data-api/v1/dataset/cff04cbb-6267-47ec-b69d-7a2315b36df3/data"}
dict_sources['Performance'] = {'url': "https://data.cms.gov/data-api/v1/dataset/8f073013-9db0-4b12-9a34-5802bdabbdfe/data"}
dict_sources['Risk'] = {'url': "https://data.cms.gov/data-api/v1/dataset/7b181182-828c-4fa4-91bd-641759f6eddd/data"}

#Check that the URL's are stored in the Keys correctly
#print(dict_sources['Participants']['url'])

#Loop for making the requests of each URL and storing the data as dataframes into the dict for use later
for key in cms_sources:
    value = f"df_{key}"
    dict_sources[key]['df'] = value
    temp_df = pd.DataFrame(requests.request("GET", dict_sources[key]['url']).json())
    dict_sources[key].update(df = temp_df)

#Check one of the DF's to see if the request function worked
dict_sources['Participants']['df'].head()

    
#response = requests.request("GET", url)

#print(response.text)
```
